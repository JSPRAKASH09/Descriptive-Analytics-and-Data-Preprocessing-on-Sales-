                                                                                               Standardization

Generally we use standardization to convert the different ranges of numerical data into a range of values, so that the mean of the data will becomes equal to Zero and standard deviation equals to 1

It helps the Machine learning model to train the data without any issues and makes easier to learn the data

Formula for standardization is  X-X(bar)/Standardizaion

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()    
```

```python
numerical_col
```

```python
standardization_data_1=data[numerical_col].copy()
```

```python
#Function to normalise the data by using the normalization Formula
def standardization_data(coll):
    return (coll-coll.mean())/coll.std()
for col in numerical_col:
    standardization_data_1[col]=standardization_data(data[col])
```

```python
standardization_data_1
```

```python
standardization_data_1.describe()
```

As we can see that the Mean and Standard Deviation for Each and every numerical column is Equals to Zero and One respectively.
So we successfully applied and normalise the data

```python
#Standard Deviation for every column is 1

print(standardization_data_1['Avg Price'].std())
print(standardization_data_1['Discount Amount'].std())
print(standardization_data_1['Discount Rate (%)'].std())
print(standardization_data_1['Net Sales Value'].std())
print(standardization_data_1['Total Sales Value'].std())
print(standardization_data_1['Volume'].std())
```

```python
#Mean for every Column is 0

print(standardization_data_1['Avg Price'].mean())
print(standardization_data_1['Discount Amount'].mean())
print(standardization_data_1['Discount Rate (%)'].mean())
print(standardization_data_1['Net Sales Value'].mean())
print(standardization_data_1['Total Sales Value'].mean())
print(standardization_data_1['Volume'].mean())
```
